### If my health and spell damage were determined by 6d4 die and my opponent's health and spell damage is determined by 4d6 die I have a feeling my opponent will win the majority of random matches.

```python
number_of_rows = 3333
my_columns = 6
enemy_columns = 4
my_health = np.random.choice([1,2,3,4], number_of_rows*my_columns).reshape(number_of_rows,my_columns)
enemy_health = np.random.choice([1,2,3,4,5,6], number_of_rows*enemy_columns).reshape(number_of_rows,enemy_columns)
my_spells = np.random.choice([1,2,3,4], number_of_rows*my_columns).reshape(number_of_rows,my_columns)
enemy_spells = np.random.choice([1,2,3,4,5,6], number_of_rows*enemy_columns).reshape(number_of_rows,enemy_columns)
```

```python
my_health = pd.DataFrame(my_health)
enemy_health = pd.DataFrame(enemy_health)
my_spells = pd.DataFrame(my_spells)
enemy_spells = pd.DataFrame(enemy_spells)
```

```python
def mage_duel(battles=1):
    # Create a dataframe that will hold number of wins.
    win_count = pd.DataFrame({'me': [0], 'enemy_mage': [0]})
    while battles > 0:
        # Set my hit points.
        my_health_tracker = np.random.choice([1,2,3,4], 6).sum()
        # Set enemy's hit points
        enemy_health_tracker = np.random.choice([1,2,3,4,5,6], 4).sum()
    
        # This will decide who attacks first each round.
        attack_first = []
    
        # The while loop for the battle
        while my_health_tracker > 0 or enemy_health_tracker > 0:
            #print(f"Your current hit points is: {my_health_tracker}")
            #print(f"The enemy mage's current hit points is: {enemy_health_tracker}")
            
            # Randomly choose who attacks first for the round.
            attack_first = np.random.choice(['me','opponent'])
            if attack_first == 'me':
                my_spell_damage = np.random.choice([1,2,3,4], 6).sum()
                #print(f'You inflicted {my_spell_damage} magic damage.')
                enemy_health_tracker = enemy_health_tracker - my_spell_damage
                if enemy_health_tracker <= 0:
                    break
                #print(f'Enemy mage has {enemy_health_tracker} hit points remaining.')
                enemy_spell_damage = np.random.choice([1,2,3,4,5,6], 4).sum()
                #print(f'The enemy mage inflicted {enemy_spell_damage} magic damage.')
                my_health_tracker = my_health_tracker - enemy_spell_damage
                if my_health_tracker <= 0:
                    break
               # print(f'You have {my_health_tracker} hit points remaining.')
            else:
                enemy_spell_damage = np.random.choice([1,2,3,4,5,6], 4).sum()
                #print(f'The enemy mage inflicted {enemy_spell_damage} magic damage.')
                my_health_tracker = my_health_tracker - enemy_spell_damage
                if my_health_tracker <= 0:
                    break
                #print(f'You have {my_health_tracker} hit points remaining.')
                my_spell_damage = np.random.choice([1,2,3,4], 6).sum()
                #print(f'You inflicted {my_spell_damage} magic damage.')
                enemy_health_tracker = enemy_health_tracker - my_spell_damage
                if enemy_health_tracker <= 0:
                    break
                #print(f'Enemy mage has {enemy_health_tracker} hit points remaining.')
        if my_health_tracker <= 0:
            win_count.enemy_mage += 1
            print('The last thing you see is the enemy mage standing over your body.')
        else:
            win_count.me += 1
            print("You stand over your opponent's lifeless body.")
        battles -= 1
    print(win_count)
```

```python
mage_duel(100)
```

## Takeaway

It seems as if throwing 6 four-sided die returns, on average, higher sums than throwing 4 six-sided die.
